```python
women_df = df[df['department'].str.lower().str.contains('women', na=False)]


scent_ratings = (
    women_df.groupby('scents')['item_rating']
    .mean()
    .sort_values(ascending=False)
)

most_popular_scent = scent_ratings.idxmax()
highest_rating = scent_ratings.max()

print(f"The most popular scent among women is '{most_popular_scent}' with an average rating of {highest_rating:.2f}")
```
